## Replace "na" values for vote_count and vote_percentage

```python
df['vote_count'] = df['vote_count'].replace("na", 0)
df['vote_percentage'] = df['vote_percentage'].replace("na", 0)
df['year'] = df['year'].apply(np.int)
df['vote_count'] = df['vote_count'].apply(np.int)
df['vote_percentage'] = df['vote_percentage'].apply(np.float)

df.sample(10)
print(df.dtypes)
```
