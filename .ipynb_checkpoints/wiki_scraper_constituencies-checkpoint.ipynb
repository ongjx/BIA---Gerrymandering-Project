```python
import requests
from bs4 import BeautifulSoup as bs
import re

def requester(url):
    res = requests.get(url)
    code = res.text
    soup = bs(code,'html.parser')
    return soup
url = 'https://en.wikipedia.org/wiki/Planning_Areas_of_Singapore'
soup = requester(url)
k = soup.find('table', attrs={'class': 'wikitable'})
# print(k.prettify())

links = []
for i in k.find_all('a'):
    links.append(i.get('href'))

links = links[1:]

for i in range(len(links)):
    base = 'https://en.wikipedia.org'
    links[i] = base + links[i]

a_dict = {}
for link in links:
    name = link.split('/')[-1]
    name = re.sub(r'[,_ ]',' ',name)
    soup = requester(link)
    k = soup.find('table', attrs={'class': 'infobox'})

    for i in k.find_all('tr'):
        if 'Constituencies' in i.getText() or "Constituency" in i.getText():
            constituencies = i.getText().split('\n')
            print(i.getText())
            if name not in a_dict:
                a_dict[name] = constituencies
pprint.pprint(a_dict)
```

```python
import pprint

b_dict = {}
check_list = []

def get_real_name(key):
    if 'Singapore' in key:
        index = key.find('S')
#         print(index)
        key = key[:index - 1]
    return key

def clean_value(list_values):
    number_constituencies = len(list_values)
    
    if list_values[number_constituencies - 1] == '':
        list_values = list_values[:number_constituencies - 1]
        
    if list_values[0] == 'Constituencies' or list_values[0] == 'Constituency':
        list_values = list_values[1:]
    
    if 'Constituencies' in list_values[0]:
        list_values[0] = list_values[0][14:]
                
        
    return list_values

for key in a_dict:
    value = a_dict[key]
    b_key = get_real_name(key)
    b_dict[b_key] = clean_value(value)
    
    
pprint.pprint(b_dict)
```
